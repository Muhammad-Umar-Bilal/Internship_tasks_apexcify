```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("sales_data.csv")

# Step 2: Show dataset
print("Dataset:\n", df)

# Step 3: Plot line chart
plt.figure(figsize=(10,5))
plt.plot(df["Month"], df["Sales"], marker='o', linestyle='-', color='b')

# Step 4: Labels and title
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Month"], df["Sales"], color="skyblue", edgecolor="black")

# Step 3: Labels and title
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend (Bar Chart)")
plt.xticks(rotation=45)
plt.show()
```
