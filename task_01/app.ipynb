```python
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv("dataset.csv")

# Step 2: Show first 5 rows
print("Dataset Preview:\n", df.head())

```

```python
# Step 3: Calculate average marks of each student
df["Average"] = df[["Math", "English", "Science"]].mean(axis=1)
```

```python
# Step 4: Find student with highest average
top_student = df.loc[df["Average"].idxmax()]
```

```python
print("\nStudent Averages (first 10):\n", df[["Name", "Math", "English", "Science", "Average"]].head(10))
print("\nTop Student:\n", top_student)
```

```python
import matplotlib.pyplot as plt

# Plot average marks of all students
plt.figure(figsize=(12,6))
plt.bar(df["Name"], df["Average"], color="skyblue")
plt.xticks(rotation=90)
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Average Marks of Students")
plt.show()
```
